```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df1 = pd.read_csv('E:\MODUL\DATASET-DATASCIENCE-20210929T070904Z-001\DATASET-DATASCIENCE\GMV Kasir.csv')
df1.head()
```

```python
df2 = pd.read_csv('E:\MODUL\DATASET-DATASCIENCE-20210929T070904Z-001\DATASET-DATASCIENCE\datamerchant.csv')
df2.head()
```

```python
df = pd.merge(df1, df2, how='inner')
df
```

```python
df_group = df.groupby(['kategori']).agg({'jumlah_transaksi': 'sum', 'GMV': 'sum'})
df_group
```

```python
df_group['jumlah_transaksi'] = (df_group['jumlah_transaksi'] / df['jumlah_transaksi'].sum()) * 100
df_group['GMV'] = (df_group['GMV'] / df['GMV'].sum()) * 100
df_group
```

```python
ax = df_group.plot.bar(rot=45)
```
